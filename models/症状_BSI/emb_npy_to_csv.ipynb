```python
import numpy as np
import pandas as pd

# 加载 .npy 文件
npy_file = "E:/comorbidity/models/症状_BSI/emb-chinese-mentalbert_BSI.npy"  # 替换为你的 .npy 文件路径

data = np.load(npy_file)

# 将 numpy 数组转换为 pandas DataFrame
df = pd.DataFrame(data)

# 保存为 .csv 文件
csv_file = 'E:/comorbidity/models/症状_BSI/emb-cn-mentalbert_BSI.csv'  # 替换为你想要保存的 .csv 文件路径
df.to_csv(csv_file, index=False, encoding='utf-8-sig')  # index=False 让 DataFrame 保存时不包含行索引
```
