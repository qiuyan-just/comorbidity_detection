# 发帖强度=总发微博数
# 互动率 = 转评赞总和 / 总发微博数
# 粉丝关注比 = 粉丝数 / 关注数（反映用户被关注的 “性价比”，比值 > 1 说明影响力强于主动社交）

```python
print(df['转评赞'].dtype)
print(df['总发微博数'].dtype)

# 尝试将列转换为数值类型
df['转评赞'] = pd.to_numeric(df['转评赞'], errors='coerce')
df['总发微博数'] = pd.to_numeric(df['总发微博数'], errors='coerce')

# 再次检查数据类型
print(df['转评赞'].dtype)
print(df['总发微博数'].dtype)
```

```python
import pandas as pd

numeric_cols = ['粉丝数', '关注数', '转评赞', '总发微博数']

# 批量转换为数值类型，无效值转换为NaN
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # errors='coerce'将无效值转为NaN
```

```python
import pandas as pd

# 假设数据已加载到DataFrame df中
# df = pd.read_csv('your_data.csv')

# 计算发帖强度（总发微博数）
# 注意：如果总发微博数已经在数据中，可直接使用df['总发微博数']
# 这里假设总发微博数需要计算（如每个用户的帖子计数）
# 如果已经有总发微博数列，可以跳过这一步
df['发帖强度'] = df['总发微博数']

# 计算互动率（转评赞总和 / 总发微博数）
df['互动率'] = df['转评赞'] / df['总发微博数']

# 计算粉丝关注比（粉丝数 / 关注数）
# 添加一个小的常数避免除零错误
df['粉丝关注比'] = df['粉丝数'] / (df['关注数'] + 0.001)  # 添加0.001避免除零

# 标记影响力强于主动社交的用户（粉丝关注比 > 1）
df['高影响力用户'] = df['粉丝关注比'] > 1

# 查看结果
print(df[['序号', '发帖强度', '互动率',  '粉丝关注比', '高影响力用户']])
```

```python
df[['序号', '发帖强度', '互动率',  '粉丝关注比', '高影响力用户']].to_excel('用户社交特征.xlsx', index=False)
```
