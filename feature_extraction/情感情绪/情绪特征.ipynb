```python
from cnsenti import Sentiment
#隶属于Sentiment类，可对文本text中的正、负面词进行统计。
#默认使用Hownet词典，后面会讲到如何导入自定义正、负情感txt词典文件。这里以默认hownet词典进行统计。
senti = Sentiment()
test_text= '我好开心啊，非常非常非常高兴！今天我得了一百分，我很兴奋开心，愉快，开心'
result = senti.sentiment_count(test_text)
print(result)
```

```python
from cnsenti import Emotion
#emotion_count(text)y用于统计文本中各种情绪形容词出现的词语数。使用大连理工大学情感本体库词典，
#支持七种情绪统计(好、乐、哀、怒、惧、恶、惊)。
emotion = Emotion()
test_text = '我好开心啊，非常非常非常高兴！今天我得了一百分，我很兴奋开心，愉快，开心'
result = emotion.emotion_count(test_text)
print(result)
```

```python
###sentiment：情感，emotion：情绪
##这部分的结果也不用，也是统计性的分析
import csv

def process_texts(input_file, output_file):
    senti = Sentiment()
    
    with open(input_file, 'r', encoding='utf-8') as infile, \
         open(output_file, 'w', newline='', encoding='utf-8') as outfile:

        writer = csv.writer(outfile)
        # 写入CSV头部
        writer.writerow([
            'doc_id',
            'count_words', 'count_sentences', 'count_pos', 'count_neg',
            'calculate_words', 'calculate_sentences', 'calculate_pos', 'calculate_neg'
        ])

        for doc_id, line in enumerate(infile, 1):
            text = line.strip()
            if not text:
                continue

            try:
                res_count = senti.sentiment_count(text)
                res_calc = senti.sentiment_calculate(text)
            except Exception as e:
                print(f"处理文档 {doc_id} 时出错：{str(e)}")
                continue

            # 提取结果（请根据实际字典键名调整）
            row = [
                doc_id,
                res_count.get('words', 0),
                res_count.get('sentences', 0),
                res_count.get('pos', 0),
                res_count.get('neg', 0),
                res_calc.get('words', 0),
                res_calc.get('sentences', 0),  # 注意键名可能需要调整
                res_calc.get('pos', 0.0),
                res_calc.get('neg', 0.0)
            ]
            
            writer.writerow(row)

# 使用示例
# process_texts("C:/Users/16437/Downloads/BERTopic-Tutorial-main/BERTopic-Tutorial-main/分词/output.txt", 'output—cnsenti-情感.csv')

process_texts("E:/comorbidity/data/output.txt", 'E:/comorbidity/feature_extraction/情感情绪/cnsenti_情感_results.csv')
```

```python
from cnsenti import Emotion
import csv
from tqdm import tqdm

def process_emotions(input_file, output_file):
    emotion = Emotion()
    
    # 先读取文件获取总行数（用于进度条）
    with open(input_file, 'r', encoding='utf-8') as f:
        total_lines = sum(1 for _ in f)
    
    with open(input_file, 'r', encoding='utf-8') as infile, \
         open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        
        writer = csv.writer(outfile)
        # 写入CSV头部
        writer.writerow([
            'doc_id',
            'words',
            'sentences',
            '好', '乐', '哀', '怒', '惧', '恶', '惊'
        ])

        # 使用tqdm创建进度条
        for doc_id, line in enumerate(tqdm(infile, total=total_lines, desc="Processing"), 1):
            text = line.strip()
            if not text:  # 跳过空行
                continue

            try:
                result = emotion.emotion_count(text)
            except Exception as e:
                print(f"\n处理文档 {doc_id} 时出错：{str(e)}")
                continue

            # 提取结果
            row = [
                doc_id,
                result.get('words', 0),
                result.get('sentences', 0),
                result.get('好', 0),
                result.get('乐', 0),
                result.get('哀', 0),
                result.get('怒', 0),
                result.get('惧', 0),
                result.get('恶', 0),
                result.get('惊', 0)
            ]
            
            writer.writerow(row)

# 使用示例
# process_emotions("C:/Users/16437/Downloads/BERTopic-Tutorial-main/BERTopic-Tutorial-main/分词/output.txt", 'emotion_results.csv')

process_emotions("E:/comorbidity/data/output.txt", 'E:/comorbidity/feature_extraction/情感情绪/cnsenti情绪结果.csv')
```
